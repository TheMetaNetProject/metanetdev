```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)
```

```python
# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75)
```

```python
# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)
```

```python
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

```python
n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
```

## This is the GMR index query

```python
from elasticsearch import Elasticsearch
es = Elasticsearch()
```

```python
resp = es.search(
    index='lms-test2',
    body={
      "query": {
        "term": {
          "target-concept": "control_of_guns"
        }
      },
      "size": 100,
      "aggs": {
        "perspective": {
          "terms": {
            "field": "perspective"
          },
          "aggs": {
            "source-concept": {
              "terms": {
                "field": "source-concepts",
                "size": 25
              }
            }
          }
        }
      }
    }
)
```

```python
resp
```

```python
aggs = resp['aggregations']
```

```python
index = np.arange(10)
```

```python
index
```

```python
io, go =[b['source-concept']['buckets'][:10] 
         for b in aggs['perspective']['buckets']]
```

```python
go
```

```python
np.array([k['doc_count'] for k in io])
```

```python
fig, ax = plt.subplots()

rects1 = plt.bar(index,
                 [k['doc_count'] for k in io], 
                 bar_width,
                 alpha=opacity,
                 color='b',
                 error_kw=error_config,
                 label='Indiv. Oversight')

rects2 = plt.bar(index + bar_width, 
                 [k['doc_count'] for k in go], 
                 bar_width,
                 alpha=opacity,
                 color='r',
                 error_kw=error_config,
                 label='Govt. Oversight')

plt.xlabel('Source Concept')
plt.ylabel('Counts')
plt.title('Gun Control')
# plt.xticks(index + bar_width, ('A', 'B', 'C', 'D', 'E'))
```
